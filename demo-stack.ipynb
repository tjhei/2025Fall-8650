```python
# see chapter 9 for classes!

class ListItem:
    def __init__(self, x, next):
        self.value = x
        self.next = next
        
class MyStack:
    """ A Stack implemented using a (singly) linked list """
    def __init__(self):
        self.top = None
        
    def push(self, x):
        newitem = ListItem(x, self.top)
        self.top = newitem
        
    def pop(self):
        # remove the top item and return it
        tmp = self.top.value
        self.top = self.top.next
        return tmp
    
    def print(self):
        x = self.top
        while x != None:
            print(x.value, end=" ")
            x = x.next
        print()
    
s = MyStack()
s.push(3)
s.push(4)
#print(s.top.value)
s.push(5)
s.print()

#print(s.top.value, s.top.next.value, s.top.next.next.value)
#print (s, s.top, s.top.value, s.top.next)
#print (s.top.value, s.top.next.value)
#print(s.pop())
#print(s.pop())
```
